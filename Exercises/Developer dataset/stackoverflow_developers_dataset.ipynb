# Stack Overflow Developer's Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('developer_dataset.csv', parse_dates=[1], index_col='RespondentID', na_values='', dtype={18:str, 19:str, 20:str})
```

```python
col_set = ['DevType', 'LanguageWorkedWith', 'LanguageDesireNextYear', 'DatabaseWorkedWith', 'DatabaseDesireNextYear', 'PlatformWorkedWith', 'PlatformDesireNextYear', 'NEWJobHunt', 'NEWJobHuntResearch']
for col in col_set:
    df[col] = df[col].fillna('null')
    df[col] = df[col].str.split(';', regex=True) 
    df[col] = df[col].map(set)
```

```python
pool = {}
col_jobhunt = ['NEWJobHunt', 'NEWJobHuntResearch']
col_pool = col_set + col_jobhunt
for col in col_pool:
    items_set = set().union(*list(df[col]))
    pool[col] = list(items_set)
```

```python
dicts = {}
for col in col_pool:
    dicts[col] = { pool[col][i]:i for i in range(len(pool[col])) }
```
